# 결정트리

결정트리는 매우 쉽고 유연하게 적용
스케일링이나 정규화 등의 사전 가공의 영향이 매우 적다

단점 : 복잡한 규칙, overfitting이 발생 → 앙상블 기법에선 장점

overfitting을 줄이기 위해 max depth나 min_sample_split를 사용

randomforest(앙상블 기법)를 위해 배워야함



###### 학습방법

규칙을 학습을 통해 자동을 분류(IF-Else 기반)



#### 균일도 기반 규칙조건

1. 정보 이득 : 1 - 엔트로피 =   정보 이득 ↑  →  정보 균일도 ↑

2. 지니계수 : 불평등지수 = 지니계수↑ → 정보 균일도 ↑



##### Decusuib Tree parameter

max_depth -트리 깊이 규정 - 깊이 규정x 시 overfitting이 발생

max_features - 피처개수 조정

min_samples_split - 최소 분할 데이터수 규정 ex) 4로 설정하면 sample 수가 4이상만 분할 가능

min_samples_leaf - 말단 노드가 되기 위한 최소한의 데이터 수 ex) 4로 설정하면 말단 노드의 sample 수가 4가 된다.



##### 결정트리 실행

1.Graphviz파일 다운

2.dot.exe파일 경로 설정

3.cmd창에서 dot -version 확인



feature를 자동으로 정하여 분류

가장 많은 class가 대표class로 지정



feature_importances 통해 중요한 feature를 학습하여 분류 시 중요 feature를 기준으로 먼저 분리



##### 결정 트리 과적합(overfitting)

영역을 너무 잘게 나누면 애매한 값이 생길 수 있다

so, 파라매터를 이용하여 적당히 나눈 후 애매한 값이 생기지 않게 한다.

다만 오류 값은 늘어남.

# 앙상블

여러 개의 분류기를 생성하고 그 예측을 결합하여 최종 예측

유형: 보팅(voting), 배깅(bagging), 부스팅(boosting), 스태킹(stacking)

### 보팅

투표를통해 최종 예측

다른 알고리즘



1. 하드 보팅 : 다수결

(디폴트값)



2. 소프트 보팅 : 확률 평균

예측성능 상대적 우수



### 배깅

ex) 랜덤 포레스트 - 여러개의 결정 트리

같은 알고리즘

데이터 샘플링을 서로 다르게 ex) 데이터를 쪼갬



랜덤포레스트 

bootstrapping : 데이터를  중첩되게 나누어서 여러 세트를 만든다. (= 데이터를 중복해서 쪼갬)



##### 하이퍼파라미터(랜덤포레스트에서 사용되는 파라미터)

1.n_estimators : 결정트리의 개수 지정

2.max_fe

3.max_dep

4.min_sam

등등



### 부스팅

ex) 에이다 부스팅, 그래디언트 부스팅, XGBoost, LightGBM
