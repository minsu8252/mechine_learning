```python
import requests
from bs4 import BeautifulSoup

res = requests.get('http://corners.gmarket.co.kr/BestSellers?viewType=C&largeCategoryCode=100000002')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('#gBestWrap > div > div:nth-child(5) > div:nth-child(3) > ul a.itemname')
# print(items)
for index,item in enumerate(items) :
    print( str(index +1) + '.', item.get_text())
```

```python
import openpyxl

def write_excel_template(filename, sheetname, listdata) :
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions['B'].width = 100
    excel_sheet.column_dimensions['C'].width = 20
    excel_sheet.column_dimensions['D'].width = 100
    excel_sheet.append(['랭킹', '제목','가격', '링크'])
    
    if sheetname !='':
        excel_sheet.title = sheetname
        
    for item in listdata :
        excel_sheet.append(item)
    excel_file.save(filename)
    excel_file.close()
```

```python
import requests
from bs4 import BeautifulSoup

product_lists = list()

res = requests.get('http://corners.gmarket.co.kr/BestSellers?viewType=C&largeCategoryCode=100000002')
soup = BeautifulSoup(res.content, 'html.parser')

data = soup.select('div.best-list')
for item in data:
    product_name = item.select_one('a.itemname')
    product_price = item.select_one('div.s-price strong span')
#     product_link = item['href']
    product_info = [product_name.get_text(), product_price.get_text()]
    product_lists.append(product_info)
```
