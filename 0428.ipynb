# 결정트리

결정트리는 매우 쉽고 유연하게 적용
스케일링이나 정규화 등의 사전 가공의 영향이 매우 적다

단점 : 복잡한 규칙, overfitting이 발생 → 앙상블 기법에선 장점

overfitting을 줄이기 위해 max depth나 min_sample_split를 사용

randomforest(앙상블 기법)를 위해 배워야함



###### 학습방법

규칙을 학습을 통해 자동을 분류(IF-Else 기반)



#### 균일도 기반 규칙조건

1. 정보 이득 : 1 - 엔트로피 =   정보 이득 ↑  →  정보 균일도 ↑

2. 지니계수 : 불평등지수 = 지니계수↑ → 정보 균일도 ↑



##### Decusuib Tree parameter

max_depth -트리 깊이 규정 - 깊이 규정x 시 overfitting이 발생

max_features - 피처개수 조정

min_samples_split - 최소 분할 데이터수 규정 ex) 4로 설정하면 sample 수가 4이상만 분할 가능

min_samples_leaf - 말단 노드가 되기 위한 최소한의 데이터 수 ex) 4로 설정하면 말단 노드의 sample 수가 4가 된다.

# 서포트 벡터 머신(SVM)

분류, 회귀에 사용

데이터 집합을 선형으로 나눔



감마 : 결정 경계를 더 세밀하게 나눔 / 값이 클수록 연산↑

C  : 벌점을 부여 / 값이 크면 over fitting / 작으면 under fitting / 적절하게 부여



```python
from sklearn import datasets
iris = datasets.load_iris()
x = iris.data
y = iris.target
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.5)
```

```python
from sklearn import svm
clf = svm.SVC(gamma=0.005, C= 100)
clf.fit(x_train, y_train)
predictions = clf.predict(x_test)
print(y_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```

# kNN

새로운 데이터를 기존 데이터와의 거리를 구해서 가까운 집합으로 분류

k 값은 몇 개의 데이터와 거리를 계산할지 규정

k값 ↑ → 메모리, 계산 시간 ↑

학습, 준비 시간  필요X

새로운 데이터를 넣을 시 따로 학습하지 않기 때문에 k값을 달리 설정해도 가능하다.

```python
from sklearn.datasets import load_iris
iris = load_iris()
#print(iris.data)
from sklearn.model_selection import train_test_split
X = iris.data
y = iris.target
# (80:20)으로 분할한다. 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
scores = metrics.accuracy_score(y_test, y_pred)
scores

X_train
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
#0 = setosa, 1=versicolor, 2=virginica
classes = {0:'setosa',1:'versicolor',2:'virginica'}
# 아직 보지 못한 새로운 데이터를 제시해보자. 
x_new = [[3,4,5,2],
[5,4,2,2]]
y_predict = knn.predict(x_new)
print(classes[y_predict[0]])
print(classes[y_predict[1]])
```
