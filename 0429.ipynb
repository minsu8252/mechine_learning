# 회귀(Regression)

연속적인 숫자를 예측에 강점

독립 변수 뿐만 아니라 회귀 계수를 알아야  예측이 가능하다

그래서 머신러닝의 회귀 예측은 학습을 통해 최적의 회귀 계수를 찾아내야한다.

y = w1 x1 + w2 x2 + ..... + wn xn

w: 회귀계수

x: 독립변수 (feature)



##### 유형

회귀 계수에 따라 나뉨

선형 회귀 : feature 1개(단일 회귀) / 거의 없다.

1. 일반 선형 회귀 : 규제 X  / RSS(비용함수)값을 최소화할 수 있도록 회귀계수를 최적화(optimize)

   - 경사 하강법(Gradient Descent): 미분한 비용함수에서 기울기가 최소인 점이 RSS(비용함수)의 최소점
     - 전 구간을 검색하기 때문에 local minima에 빠지지 않고 global minima을 찾는다.
     - stochastic gradient를 많이  사용
     - 프로세스 
       - 1. w 임의로 설정 후 비용함수 계산
         2.  조금씩 w를 바꾸면서 계산
         3. 비용함수 값이 감소했으면 step2 반복, 비용함수가 감소하지 않으면 반복 중지

2. 릿지(Ridge) : L2규제(값이 작은거 규제)

3. 라쏘(Lasso) : L1규제(값이 큰거 규제)

4. 엘라스틱넷(ElasticNet) : L1 + L2 규제

5. 로지스틱 회귀(): 회귀 X  / 분류 O
